# Preparing Data For Training


```python
data['TotalCharges'].head()
```

```python
# convert to numeric
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
```

```python
# datatypes of the columns
data.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Encoding categorical variables using LabelEncoder
categorical_cols = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod', 'Churn'] # Changed 'Paperless' to 'PaperlessBilling'
label_encoder = LabelEncoder()
for col in categorical_cols:
    data[col] = label_encoder.fit_transform(data[col])
```

```python
data
```

```python
# finding data with NaN
data.isnull().sum()
```

```python
# fill missing value with the mean
data['TotalCharges'].fillna(data['TotalCharges'].mean(), inplace=True)
```

```python
# drop customerID
data.drop('customerID', axis=1, inplace=True)
```

```python
y_target = data['Churn']
```

```python
# correlation
data.corr()
```

```python
# plot correlation
plt.figure(figsize=(10,5))
sns.heatmap(data.corr(),annot=True)
plt.show()
```

```python
# drop churn and customer id
data = data.drop(['Churn'], axis=1)
```

# Building a Logistic Regresstion model

```python


from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, roc_curve, roc_auc_score, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# splitting the data
x_train, x_test, y_train, y_test = train_test_split(data, y_target, test_size=0.2, random_state=42)
```

```python
# Train Logistic Regression model
model = LogisticRegression(random_state=42, max_iter=1000)
model.fit(x_train, y_train)
```

```python

# Predict and evaluate
y_pred = model.predict(x_test)
y_proba = model.predict_proba(x_test)[:, 1]
```

```python


# Classification report
report = classification_report(y_test, y_pred,) # target_names=target_encoder.classes_)
```

```python
print(report)
```

```python
# accuracy
accuracy_score(y_test, y_pred)
```

```python
# confusion matrix
confusion_matrix(y_test, y_pred)
```

```python
coefficients = pd.DataFrame({'Feature': data.columns, 'Coefficient': model.coef_.flatten()}) # Flatten model.coef_ to 1-D
print(coefficients.sort_values(by='Coefficient', ascending=True))
```

```python
# plot confusion matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python

# Calculate and plot ROC curve and AUC score
fpr, tpr, thresholds = roc_curve(y_test, y_proba)
auc_score = roc_auc_score(y_test, y_proba)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"Logistic Regression (AUC = {auc_score:.2f})")
plt.plot([0, 1], [0, 1], "k--", label="Random Guess")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()

```
